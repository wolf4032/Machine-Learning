```python
# .index = [各種つけたい名前]　indexに名前をつける
df.index = ["Taro", "Jiro", "Hana", "Ayumi"]
print(df)
print(df.index)
```

```python
# .rename(columns={"変更前の列名":"変更後の列名"}, index={"変更前のインデックス":"変更後のインデックス"})　特定の列と行だけ名前を変更
# 引数は辞書型、キーは変更前、バリューは変更後
df = df.rename(columns = {"Gender" : "Sex"}, index = {"Taro" : "Goro"})
df
```

```python
# df["追加する列の名前"] = [各種列の値]　列の追加
# 前処理の過程でよく用いる
# df["追加する列の名前"] =　0　などとすると、追加した列の全ての値を0として追加できる
df["City"] = ["Tokyo", "Kyoto", "Okinawa", "Hakata"]
df
```

```python
# .drop(columns = "削除したい列名")　列を削除
# 前処理の過程でよく用いる
# axis=0（デフォルト）：行を削除する
# axis=1：列を削除する
# 今回のように、columns = と指定すると、axisの記述がなくても列を削除してくれる？
df = df.drop(columns = "City")
df
```

```python
# データフレームの参照
```

```python
# df["列名"]　df.列名　一列のデータを参照
df["Age"]
```

```python
# df[[参照したい各種列名]]　(df.[参照したい各種列名]ではエラーになる)複数列同時に参照　引数はリスト
df[["Age", "Sex"]]
```

```python
# .loc["インデックス", "ヘッダー"]　インデックスとヘッダーを同時に指定して特定のデータを取り出す
# 超重要
# .loc[条件式]などにすると、条件に合う行とその列を全て表示する
df.loc["Goro", "Age"]
```

```python
# .loc[:, [各種ヘッダー名]]任意の行の複数の列を取り出す
df.loc[:, ["Age", "Sex"]]
```

```python
# df[df["列名"] 条件式]　条件を満たすデータ(行)を取り出す
# df["列名"]でその列の値を意味する？
# データフレームを関数とすると列名は関数に対する引数のように特定の入力に対して特定の出力がされるように？

# 厳密には条件に対してTrueのものを返す文
# df["列名"] 条件式とすると、各行がTrueかFalseかを示す一覧が表示される
df[df["Age"] > 12]
```

```python
# locで条件を指定する
df.loc[df["Age"] > 12, ["Height", "Weight"]]
# Ageが12以上のデータのHeightとWeightを取り出す
```

```python
# sort_values("列名", ascending = True/False)　データを並び替える
# ascendingは昇順という意味、昇順であれば、引数は列名だけでいい
df.sort_values("Age", ascending = True)
```

```python
# 複数の列を指定して並び替える場合は、並べ替えの優先度が高い順に列名をリストで渡す
df.sort_values(["Sex", "Age"])
```

```python
# df[列名].value_counts() 列内の各値の出現回数を求める  多い順に並ぶ
#..._counts().keys()で探す対象(keys()のほかに、indexも使えるかも？)、..._counts().valuesでそれぞれの件数
df["Sex"].value_counts()
```

```python
# .groupby("列名").集計するもの　その列の値が等しいものの集計結果を列の値ごとに出す
# 列名が複数ならリストで渡す、左から順に優先順位が高い
# 各性別の性別以外の列の平均値を求めるなど
df.groupby("Sex").mean()
```

```python
# .演算子or関数　基本統計量を出す
# 特定の列に対して行うには、.loc[:, "列名"]が必要
# max()　最大値
print(df.loc[:, "Weight"].max())
# median()　中央値
print(df.loc[:, "Weight"].median())
# min()　最小値
print(df.loc[:, "Weight"].min())
```

```python
# 数学的な演算　特定の列全体に同様の編集、変更を行う
df["Age"] += 1
df
```

```python
# describe()　基本統計量をまとめて出す
# 特定の列に対して行うには、.loc[:, "列名"]が必要
print(df.loc[:, "Weight"].describe())
```

```python
# 欠損値：値の存在しないデータ
# NaNと表される
# 欠損値が含まれると分析が困難、何かしらの対処が必要

# 対処法
# １．欠損値を含む行を削除する
# ２．欠損値を平均値や中央値などの別の値で置き換える
```

```python
# 欠損したデータフレーム
# 欠損値はnp.nan
df = pd.DataFrame({"Age" : [12, 14, 12, 18],
                  "Gender": ["M", "M", "F", "F"],
                  "Height": [np.nan, 160, 150, 175],
                  "Weight": [35, 50, np.nan, 60]})
df
```

```python
# isnull()　欠損値の存在を確認　Trueが欠損値
df.isnull()
```

```python
# isnull().sum()　欠損値の数を数える　超重要
# PythonではTrueは1、Falseは0
df.isnull().sum()
```

```python
# dropna()　欠損値の行を削除
# 欠損値が全くない行はそうそうなく、かなりのデータが消えることになるため、できる限りしない
# df_dropna = df.dropna()
# df_dropna
```

```python
# fillna(置き換える後の値)　欠損値の置き換え
df_fillna = df.fillna(df.mean())
df_fillna
```
