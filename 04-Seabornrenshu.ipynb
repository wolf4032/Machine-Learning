```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips=pd.read_csv("tips/tips.csv")
```

```python
# 棒グラフ：countplot()
# 引数で与えたlistやarrayのカテゴリごとのデータ数を自動で数えて棒グラフ化
# 動画ではsns.countplot(tips.day)で説明されていたが、dayの各曜日は数値型でなく文字列型で、
# その場合はdataパラメータにデータフレームを、xパラメータに列を指定しなければ処理できないらしい
sns.countplot(data=tips, x='day')
# Matplotlibより短いコードで可視化できる
"""
left=range(4)
day=tips.day.value_counts().keys()
height=tips.day.value_counts().values
plt.bar(left,height,tick_label=day)
"""
```

```python
# 棒グラフ：barplot()
# 先の棒グラフと異なり、平均値と値のばらつき（標準偏差を表す）

# 各曜日の支払総額の平均とばらつき
# 動画内のsns.barplot(tips.day,tips.total_bill)ではだめだった、plt.show()もいらなかった
sns.barplot(data=tips, x='day', y='total_bill')
```

```python
# ヒストグラム：distplot()
# Matplotlibのヒストグラムの縦軸は件数だったが、Seabornは0～1の%表示
sns.distplot(tips.total_bill,bins=20)
"""
・distplot()はヒストグラムとヒストグラムを滑らかにした曲線を重ねて描画
・引数でkde=Falseと指定すると曲線を描画しない
・引数でhist=Falseと指定すると、ヒストグラムを描画しない
"""
```

```python
# 線形回帰プロット: regplot()
# データの傾向を反映した直線と散布図
# 動画内のsns.regplot(tips.total_bill,tips.tip)ではだめだった
sns.regplot(data=tips, x="total_bill", y="tip")
"""
xlabel、ylabelが自動で付与される
データの傾向を反映した直線（線形回帰）の直線が自動で描かれる
９５％信頼区間（直線の確信度のようなもの）が自動で描かれる
    データが少なく、確信度が低いところほど区間が広い
引数scatter=Falseにすると散布図が消える
引数fit_reg=Falseにすると直線が消える
"""
```

```python
"""
重要！Seabornで書いたプロットを並べる
復習　Matplotlibにおいてグラフを複数並べたいときに使う関数は？
    subplots()
では、subplots()によって生成されるのは？
    figureとaxes
    fig,ax = subplots(1,2)
"""
```

```python
# Seabornで描いたプロットを並べる
# Seabornのグラフを並べるとき　プロットを行うaxesの名前を引数axに代入する
fig,ax=plt.subplots(1,2,figsize=(10,5))
sns.countplot(data=tips, x="day", ax=ax[0])
sns.regplot(data=tips, x="total_bill", y="tip", ax=ax[1])
```

```python
"""
Seabornの強み1
ヒストグラムと散布図は重要な可視化
    ヒストグラム：異常値の把握、データが従う確率分布の推定
    散布部：相関の把握

ヒストグラムと散布図を１つ１つ出すのは面倒、相関係数を具体的に数値で出したい
    これらのニーズを満たす可視化がある
"""
```

```python
# 散布図行列(ペアプロット)：pairplot()
sns.pairplot(tips)
"""
全ての量的変数に対して、ヒストグラムと散布図を表示
データの分布・相関についての情報を一挙に得られる

量的変数：数字で表されている変数（データ）
        　曜日や性別の変数は文字列なので量的変数ではない
"""
"""
sns.pairplot(データフレーム名[列名])
調べる列を指定する場合の書き方、複数の列名はデータフレーム名の引数の中でリストで渡す
"""
```

```python
"""
相関係数：２つの変数の関係の値　-1～1　散布図の状態で分かる
    -1:右肩下がり　負の相関関係
    0:均等に散らばっている　相関無し
    1:右肩上がり　正の相関関係
"""
# 相関関係の可視化：heatmap()
# Pandas スタイル
# corr():相関関係を出す
# .style.background_gradient():相関の情報に色を付ける
tips.corr().style.background_gradient("summer_r")
```

```python
# 相関関係の可視化：heatmap()
# Seaborn　スタイル
# スケールバーがつく
# annot=True ヒートマップの中に値を追加
sns.heatmap(tips.corr(),cmap="summer_r")
```

```python
"""
Seabornの強み２
データの分布をみるときに「カテゴリごとに見たい」という場合がある
様々な描画方法があるが今回は以下の２種類を紹介
    swarmplot()
    boxplot()
"""
```

```python
# カテゴリごとにデータの分布を表示する：swarmplot()①
# 動画内のsns.swarmplot("day","total_bill",data=tips)はエラーだった
sns.swarmplot(data=tips, x="day", y="total_bill")
```

```python
# カテゴリごとにデータの分布を表示する：swarmplot()②
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
"""
hue（色相）
    可視化の切り口をさらに追加
    
dodge（かわす、回避する）
    Trueで、追加した切り口を重ねない
"""
```

```python
# 切り口を重ねるバージョン
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=False)
```

```python
# カテゴリごとにデータの分布を表示する（箱ひげ図）：boxplot()
# 動画内のsns.boxplot(tips.day,tips.tip,hue=tips.sex)ではエラーが出た
sns.boxplot(data=tips, x="day", y="tip", hue="sex")

"""
箱ひげ図
中央値：真ん中の線
各横線がデータの数が25%毎の区切り（色が塗ってあるところが、中央値から±25%の範囲）
黒点：外れ値
"""
```
