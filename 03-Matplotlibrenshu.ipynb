```python
plt.figure(figsize=(4,4)) # 横4inch、縦4inch
plt.plot([0,2,4,6,8],[0,6,12,24,48], label="graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title")
plt.legend() # 凡例を表示する
plt.show() # グラフの表示(jupyter notebook上では記載なくても描画される)
```

```python
x = np.linspace(0, 10, 20) # 0～10を等間隔で20個、arrangeだと10は含まない
y = x ** 2
z = (x - 1) ** 3
w = ((x - 2) * (x - 4)) ** 2
plt.plot(x, y, label = "y = x ^ 2")
plt.plot(x, z, label = "z = (x - 1) ^ 3")
plt.plot(x, w, label = "w = {(x - 2) * (x - 4)} ^ 2")
plt.legend()
plt.show()
```

```python
# 複数のグラフを別々に書く
plt.figure(figsize = (12, 4))

plt.subplot(1, 3, 1) # 1行3列の1列目
plt.plot(x, y, label = "y = x ^ 2")
plt.legend()

plt.subplot(1, 3, 2) # 1行3列の2列目
plt.plot(x, z, label = "z = (x - 1) ^ 3")
plt.legend()

plt.subplot(1, 3, 3) # 1行3列の3列目
plt.plot(x, w, label = "w = {(x - 2) * (x - 4)} ^ 2")
plt.legend()

plt.tight_layout() # グラフ同士の重なりを防ぐ
```

```python
"""
Pyplotインターフェース
    ・plt.～でひたすら書いていく
    ・これまで行ってきた方法
    ・1つのグラフを書くときに向いている
オブジェクト指向インターフェース
    ・グラフ１つ1つをオブジェクトとして管理する
    ・複数のグラフを描画するときに向いている
    """
```

```python
"""
オブジェクト指向インターフェースのグラフ構成
figure:描画領域、キャンバス全体
axes:キャンバスの一部分
x axis:x軸
y axis:y軸
label:凡例
"""
```

```python
# オブジェクト指向インターフェースの基本
```

```python
# パターン１：plt.add_subplot
fig = plt.figure(figsize = (3, 3)) # 空のfigureを作り、figと名付ける
ax = fig.add_subplot(1, 1, 1) # 1行1列のsubplotをfigの中に作成、1列目をaxと名付ける
ax.plot(x, y, label = "y = x ^ 2") # ax.plot()はaxというaxesにプロットせよという意味
ax.plot(x, z, label = "z = (x - 1)^ 3")
ax.plot(x, w, label = "w = ((x - 2)(x - 4))^2")
ax.set_xlim(0, 10) # xの最大値と最小値
ax.legend()
```

```python
"""
オブジェクト指向インターフェースでは
axesに対してtitleやxlabelを定めるときはset_が先頭につく
例：
plt.title("～")→ax.set_title("～")
plt.xlabel("～")→ax.set_xlabel("～")
"""
```

```python
# パターン２：plt.subplots
    # 一つのグラフ
fig, ax = plt.subplots(1, 1, figsize = (3, 3)) # figureとaxesを同時に生成、3inch*3inchのfugと1行1列のaxesを生成
ax.plot(x, y, label = "y=x^2")
ax.plot(x, z, label = "z=(x-1)^3")
ax.plot(x, w, label = "w=((x-2)(x-4))^2")
ax.legend()
ax.set_title("three curves")
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
```

```python
# パターン２：plt.subplots
    # １行３列のグラフ
fig, ax = plt.subplots(1, 3, figsize = (14, 4)) 
print(ax) # axはarray(行列、Numpy配列)形式であることを確認
ax[0].plot(x, y, label = "y=x^2")
ax[1].plot(x, z, label = "z=(x-1)^3")
ax[2].plot(x, w, label = "w=((x-2)(x-4))^2")
for i in range(3): # 三つのグラフを順番に描画
    ax[i].legend() # set_はつけない
    ax[i].set_xlabel("x axis")
    ax[i].set_ylabel("y axis")
plt.tight_layout()
```

```python
# パターン２：plt.subplots
    # ２行２列のaxesを作った場合
    # Numpy配列の位置引数は[行,列]で指定する
fig,ax = plt.subplots(2,2)
ax[0,0].plot(x, y, label = "y=x^2")
ax[0,0].legend()
ax[0,1].plot(x, z, label = "z=(x-1)^3")
ax[0,1].legend()
ax[1,0].plot(x, w, label = "w=((x-2)(x-4))^2")
ax[1,0].legend()
```

```python
"""
演習　Matplotlibの基本事項
問題１
    y =sin x, y = cos x, y = tan x
    -pi <= x <= pi
    figsizeは（8,6）
    線の色はplot(color = "red")等でできる
"""
```

```python
fig, ax = plt.subplots(3, 1, figsize=(8,6))
x = np.linspace(-np.pi, np.pi, 100)
ax[0].plot(x, np.sin(x), label="y=sin x", color="orange")
ax[1].plot(x, np.cos(x), label="y=cos x", color="blue")
ax[2].plot(x, np.tan(x), label="y=tan x", color="green")
for i in range(3):
    ax[i].legend()
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("y")
plt.tight_layout()
```

```python
# 機械学習における可視化でよく用いる表現形式１
"""
１．あるデータ列の各要素の数が知りたい
    棒グラフ
    「曜日」列における各曜日の個数
２．あるデータ列の各要素の全体に対する割合が知りたい
    円グラフ
    「曜日」列における各曜日の全体に対する割合
３．あるデータ列の値の分布の様子が知りたい
    ヒストグラム
    「支払総額」のデータの取りうる値の様子
４．２つのデータ列の値の分布の様子が知りたい
    散布図
    「支払総額」と「チップの額」の取りうる値・関係性
    """
```

```python
# 棒グラフ plt.bar()　数の大小を比較したい
left=range(4) # 軸の数分、角帽のx軸上の数値
day=tips.day.value_counts().keys()    # tipsデータフレームのday列の各曜日の出現数を取得し、各曜日名を返す、keys()の代わりにindexも使える？
height=tips.day.value_counts().values # 　　　　　　　　　　　''   　　　　　　　　　　　　各曜日の出現数を返す、棒の高さ
plt.bar(left,height,tick_label=day) # tick_labelでleftによって振られた数字のラベルを文字列で置き換える
```

```python
# 円グラフ　plt.pi()　数の大小を全体に対する割合で比較したい
plt.pie(tips.day.value_counts(),autopct="%.1f%%",labels=tips.day.value_counts().keys())
# autopct:
#     フォーマット指定
#     割合を「～%」としてグラフに表示させる設定
#     %.3f%%とすれば小数第３位まで表示
    
# labels:
#     要素のラベル
```

```python
# ヒストグラム　plt.hist()　データが取り得る値の様子を確認したい
plt.hist(tips.total_bill, bins=20, color="green")
plt.xlabel("total_bill")
plt.ylabel("counts")
plt.title("histogram of total_bill")
# bins:ビンの数を指定（ビン＝縦棒の数）
```

```python
# 散布図　plt.scatter()　組データの取り得る値の様子を確認したい
plt.scatter(x=tips["total_bill"],y=tips["tip"]) # plt.scatter("total_bill","tip",data=tips)でも同じ結果
plt.xlabel("total_bill")
plt.ylabel("tip")
```

```python
"""
Day5で学んだ知識を確認
１．
    一番外の要素をfigure、その中の要素をaxesという

２．
    ヒストグラムとは何を調べるための可視化表現？
        １変数の分布
    
    散布図とは何を調べるための可視化表現？
        ２変数の分布
        
    ヒストグラムと散布図を作成するメソッドは？
        ヒストグラム：hist()
            正規分布に近いほど、学習しやすいデータ
            ロングテールなデータは正規分布になる部分のみ抜き出すなどする
        散布図：sucatter()
            予測したい変数と関係のある変数を調べる
"""
```

```python
"""
演習　Matplotlibによる可視化
問題１　性別ごとにチップの額と支払総額の関係性を調べたい。適切な可視化を行え。
    　　ヒント：チップの額と支払総額のデータを性別ごとに作成せよ
          total_bill:支払総額
          tip:チップの額
"""
```

```python
fig,ax=plt.subplots(2,1,figsize=(8,8))

tips_male = tips[tips["sex"] == "Male"][["total_bill","tip"]] # tips_male = tips["sex"列の値が"Male"の行]["total_bill"と"tip"の列]
tips_female = tips[tips["sex"] == "Female"][["total_bill","tip"]]

ax[0].scatter(x=tips_male["total_bill"],y=tips_male["tip"],label="Male",color="blue") # 点の大きさはsという変数に数値を代入して指定する
ax[1].scatter(x=tips_female["total_bill"],y=tips_female["tip"],label="Female",color="red")

for i in range(2):
    ax[i].set_xlabel("total_bill")
    ax[i].set_ylabel("tips")
    ax[i].legend()
```
